```python
import pandas as pd
import numpy as np

# Charger le CSV
df = pd.read_csv("sample_data/mnist_test.csv")


# Séparer labels et images
y_test = df.iloc[:, 0].values          # labels
X_test = df.iloc[:, 1:].values         # pixels

print(X_test.shape, y_test.shape)
```

```python
# Reshape vers (N, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

# Normalisation
X_test = X_test / 255.0

print(X_test.shape)
```

```python
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical

# Charger le dataset
(_, _), (X_test, y_test) = fashion_mnist.load_data()

# Prétraitement images
X_test = X_test.reshape(-1, 28, 28, 1) / 255.0

# One-hot encoding labels
y_test = to_categorical(y_test, num_classes=10)

# 🔥 Recompiler le modèle (OBLIGATOIRE)
model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

# Évaluation
loss, acc = model.evaluate(X_test, y_test, verbose=1)

print("✅ Accuracy réelle :", acc)
```
